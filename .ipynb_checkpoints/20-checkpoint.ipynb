**Задача 2.1:** Данные о выручке сервиса представлены в российских рублях и казахстанских тенге. Приведите выручку к единой валюте — российскому рублю.

Для этого используйте датасет с информацией о курсе казахстанского тенге по отношению к российскому рублю за 2024 год — `final_tickets_tenge_df.csv`. Его можно загрузить по пути `https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')`

Значения в рублях представлено для 100 тенге.

Результаты преобразования сохраните в новый столбец `revenue_rub`.

---


```python
# Загрузка данных о курсе тенге
tenge_url = 'https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv'
tenge_df = pd.read_csv(tenge_url)

print("Данные о курсе тенге:")
print(tenge_df.head())
print(f"\nРазмер: {tenge_df.shape}")
print(f"\nСтолбцы: {tenge_df.columns.tolist()}")

# Конвертация типов данных для объединения
tenge_df['order_dt'] = pd.to_datetime(tenge_df['data'])
df['order_dt'] = pd.to_datetime(df['order_dt'])

# Объединение данных о курсе с основной таблицей
df = df.merge(tenge_df, on='order_dt', how='left')

# Расчет выручки в рублях
# Для рублей: оставляем как есть
# Для тенге: делим на 100 и умножаем на курс в рублях
df['revenue_rub'] = df.apply(
    lambda row: row['revenue'] if row['currency_code'] == 'rub' 
    else row['revenue'] / 100 * row['curs'], 
    axis=1
)

# Проверка результата
print("\nПроверка конвертации:")
print(df[df['currency_code'] != 'rub'][['currency_code', 'revenue', 'revenue_rub']].head(10))
print(f"\nРублевых транзакций: {(df['currency_code'] == 'rub').sum()}")
print(f"Тенговых транзакций: {(df['currency_code'] == 'kzt').sum()}")
```

конвертация прошла успешно и добавлена новая колонка revenue_rub

- среднее время между заказами.

После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


```python
df.head(3)
```

```python
# Сортировка данных по времени заказа
df = df.sort_values(['user_id', 'order_ts']).reset_index(drop=True)
```

```python
# Построение профиля пользователя
user_profile = df.groupby('user_id').agg(
    # Даты первого и последнего заказа
    first_order_date=('order_dt', 'first'),
    last_order_date=('order_dt', 'last'),
    
    # Инфо о первом заказе
    first_device=('device_type_canonical', 'first'),
    first_region=('region_name', 'first'),
    first_service=('service_name', 'first'),
    first_event_type=('event_type_main', 'first'),
    
    # Агрегированные метрики
    total_orders=('order_id', 'count'),
    avg_revenue=('revenue_rub', 'mean'),
    avg_tickets=('tickets_count', 'mean'),
    avg_days_between_orders=('days_since_prev', 'mean')
).reset_index()
```

```python
user_profile['is_two'] = (user_profile['total_orders'] >= 2).astype(int)
user_profile['is_five'] = (user_profile['total_orders'] >= 5).astype(int)
```

```python
print("\nПрофиль пользователя создан")
print(f"\nРазмер профиля: {user_profile.shape}")

print('\nпервые пять строк профилей')
print(user_profile.head(5))

# Статистика по профилям
print("\nСтатистика профилей:")
print(user_profile.describe())
```

#### 4.4. Корреляционный анализ количества покупок и признаков пользователя

Изучите, какие характеристики первого заказа и профиля пользователя могут быть связаны с числом покупок. Для этого используйте универсальный коэффициент корреляции `phi_k`, который позволяет анализировать как числовые, так и категориальные признаки.

Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;

```python
# Выбираем признаки для анализа
features_for_analysis = [
    'total_orders',
    'avg_revenue',
    'avg_tickets',
    'avg_days_between_orders',
    'first_event_type',
    'first_device',
    'first_region',
    'first_service',
    'first_order_weekday'
]

# Создаем копию для анализа
analysis_df = user_profile[features_for_analysis].copy()

# Проверка распределения total_orders
print("\nРаспределение total_orders:")
print(analysis_df['total_orders'].value_counts().sort_index().head(10))
print(analysis_df['total_orders'].describe())

# Визуализация распределения
plt.figure(figsize=(12, 5))
analysis_df['total_orders'].value_counts().sort_index().head(20).plot(
    kind='bar',
    rot=45,
    legend=False,
    title='Распределение количества заказов',
    figsize=(12, 5),
    color='steelblue',
    edgecolor='black'
)
plt.xlabel('Количество заказов')
plt.ylabel('Количество пользователей')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Определяем интервальные признаки
interval_cols = ['total_orders', 'avg_revenue', 'avg_tickets', 'avg_days_between_orders']

# Вычисляем корреляционную матрицу
correlation_matrix_all = analysis_df.phik_matrix(interval_cols=interval_cols)

print("\nКорреляционная матрица:")
print(correlation_matrix_all)

# Извлекаем корреляции с total_orders
correlations_with_orders = correlation_matrix_all['total_orders'].sort_values(ascending=False)
print("\nКорреляция признаков с total_orders:")
print(correlations_with_orders)

# Визуализация тепловой карты
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_all, annot=True, fmt='.3f', cmap='RdYlGn', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Корреляционная матрица phi_k (все пользователи)', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Создание сегментов
user_profile['order_segment'] = pd.cut(
    user_profile['total_orders'],
    bins=[0, 1, 4, float('inf')],
    labels=['1 заказ', '2-4 заказа', '5+ заказов'],
    right=True
)

# Статистика по сегментам
segment_stats = user_profile.groupby('order_segment', observed=True).agg(
    count=('user_id', 'count'),
    avg_orders=('total_orders', 'mean'),
    avg_revenue=('avg_revenue', 'mean'),
    avg_tickets=('avg_tickets', 'mean')
)
segment_stats['share_pct'] = (segment_stats['count'] / len(user_profile) * 100)

print("\nСтатистика по сегментам:")
print(segment_stats)
```

```python
segments = ['1 заказ', '2-4 заказа', '5+ заказов']
correlation_results = {}

for segment in segments:
    print(f"\n{segment.upper()}:")
    print("-" * 50)
    
    # Фильтруем данные по сегменту
    segment_df = user_profile[user_profile['order_segment'] == segment][features_for_analysis].copy()
    
    print(f"Количество пользователей: {len(segment_df):,}")
    
    # Проверяем вариацию в total_orders
    unique_orders = segment_df['total_orders'].nunique()
    print(f"Уникальных значений total_orders: {unique_orders}")
    
    if unique_orders <= 1:
        print("Нет вариации в total_orders - корреляционный анализ невозможен")
        correlation_results[segment] = None
        continue
    
    # Вычисляем корреляцию
    try:
        correlation_matrix_segment = segment_df.phik_matrix(interval_cols=interval_cols)
        
        # Сохраняем результаты
        correlation_results[segment] = correlation_matrix_segment
        
        # Корреляции с total_orders
        if 'total_orders' in correlation_matrix_segment.columns:
            correlations = correlation_matrix_segment['total_orders'].sort_values(ascending=False)
            print("\nКорреляция с total_orders:")
            print(correlations)
        else:
            print("\nСтолбец total_orders отсутствует в матрице корреляций")
    
    except Exception as e:
        print(f"\nОшибка при расчете корреляции: {e}")
        correlation_results[segment] = None
```

```python
# Фильтруем только сегменты с результатами корреляции
valid_segments = [seg for seg in segments if correlation_results[seg] is not None]

# Создаем субплоты только для валидных сегментов
fig, axes = plt.subplots(1, len(valid_segments), figsize=(20, 6))

# Если только один валидный сегмент, axes не будет массивом
if len(valid_segments) == 1:
    axes = [axes]

for idx, segment in enumerate(valid_segments):
    ax = axes[idx]
    sns.heatmap(correlation_results[segment], annot=True, fmt='.3f', 
                cmap='RdYlGn', center=0, square=True, linewidths=1, 
                cbar_kws={"shrink": 0.8}, ax=ax)
    ax.set_title(f'{segment}\n(n={segment_stats.loc[segment, "count"]:,})', 
                 fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Собираем корреляции в одну таблицу (только для сегментов с результатами)
comparison_data = {
    'Все пользователи': correlation_matrix_all['total_orders']
}

# Добавляем только сегменты с результатами
for segment in ['1 заказ', '2-4 заказа', '5+ заказов']:
    if correlation_results[segment] is not None and 'total_orders' in correlation_results[segment].columns:
        comparison_data[segment] = correlation_results[segment]['total_orders']

comparison_df = pd.DataFrame(comparison_data)

print("\nСравнение корреляций с total_orders:")
print(comparison_df.sort_values('Все пользователи', ascending=False))

# Визуализация сравнения
comparison_df_plot = comparison_df.drop('total_orders', errors='ignore')  # Убираем автокорреляцию

plt.figure(figsize=(14, 6))
comparison_df_plot.plot(kind='bar', rot=45, figsize=(14, 6), 
                        alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.ylabel('Корреляция (phi_k)', fontsize=12)
plt.xlabel('Признак', fontsize=12)
plt.title('Сравнение корреляций с total_orders по сегментам', 
          fontsize=14, fontweight='bold')
plt.legend(title='Сегмент', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
